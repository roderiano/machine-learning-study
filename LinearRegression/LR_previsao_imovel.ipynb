```python
# Importando as bibliotecas necessrias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Dados de exemplo
area = np.array([50, 60, 70, 80, 90, 100, 110, 120]).reshape(-1, 1)  # rea da casa em metros quadrados
num_quartos = np.array([2, 2, 3, 3, 4, 4, 5, 5])  # Nmero de quartos
num_banheiros = np.array([1, 1, 2, 2, 3, 3, 4, 4])  # Nmero de banheiros
preco = np.array([300000, 350000, 400000, 450000, 500000, 550000, 600000, 650000])  # Preo da casa em reais

# Combinando as variveis independentes em uma nica matriz
X = np.column_stack((area, num_quartos, num_banheiros))

# Visualizando os dados
fig = plt.figure(figsize=(12, 4))

ax1 = fig.add_subplot(131, projection='3d')
ax1.scatter(area, num_quartos, preco, color='blue')
ax1.set_title('rea vs Nmero de Quartos vs Preo')
ax1.set_xlabel('rea (m)')
ax1.set_ylabel('Nmero de Quartos')
ax1.set_zlabel('Preo (R$)')

ax2 = fig.add_subplot(132, projection='3d')
ax2.scatter(area, num_banheiros, preco, color='red')
ax2.set_title('rea vs Nmero de Banheiros vs Preo')
ax2.set_xlabel('rea (m)')
ax2.set_ylabel('Nmero de Banheiros')
ax2.set_zlabel('Preo (R$)')

ax3 = fig.add_subplot(133, projection='3d')
ax3.scatter(num_quartos, num_banheiros, preco, color='green')
ax3.set_title('Nmero de Quartos vs Nmero de Banheiros vs Preo')
ax3.set_xlabel('Nmero de Quartos')
ax3.set_ylabel('Nmero de Banheiros')
ax3.set_zlabel('Preo (R$)')

plt.tight_layout()
plt.show()

# Treinando o modelo de regresso linear
modelo = LinearRegression()
modelo.fit(X, preco)

# Fazendo previses para uma nova casa
nova_casa = np.array([[95, 3, 2]])  # Nova casa com 95m, 3 quartos e 2 banheiros
preco_predito = modelo.predict(nova_casa)
print('Preo predito para a nova casa:', preco_predito[0])
```
