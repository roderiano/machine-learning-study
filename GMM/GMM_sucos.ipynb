```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

# definir uma semente aleatria para garantir a reprodutibilidade dos resultados
np.random.seed(60)

# Criao de dados fictcios para a acidez do suco de frutas em duas estaes do ano: Vero e Primavera.
# A acidez do suco de frutas em diferentes estaes do ano pode seguir distribuies diferentes.
# Vamos assumir que a acidez do suco de frutas no Vero segue uma distribuio com mdia 5 e desvio padro 1,
# enquanto a acidez do suco de frutas na Primavera segue uma distribuio com mdia -5 e desvio padro 1.

data_summer = np.random.normal(5, scale=1, size=700)  # 700 pontos com mdia 5 e desvio padro 1 para o Vero
data_spring = np.random.normal(-5, scale=1, size=300)  # 300 pontos com mdia -5 e desvio padro 1 para a Primavera

data = np.concatenate([data_summer, data_spring])  # Combina os dados de Vero e Primavera

# Visualiza o conjunto de dados original com um histograma
plt.hist(data, bins=30, density=True)
plt.title("Acidez do suco de frutas original")
plt.show()

# Cria e treina um modelo de Mistura Gaussiana com dois componentes (dois picos) para modelar a acidez do suco de frutas
gmm = GaussianMixture(n_components=2)
gmm.fit(data.reshape(-1, 1))  # Ajusta o modelo aos dados

# Usa o modelo treinado para gerar 1000 novos pontos de dados simulando o suco de frutas da estao Primavera
samples = gmm.sample(1000)
sample_spring = samples[0]

# Plota um histograma comparando a acidez do suco de frutas original e a acidez do suco de frutas simulado para a estao Primavera
plt.hist(sample_spring, bins=30, density=True, alpha=.5, color='green', label='Suco Primavera')
plt.hist(data, bins=30, density=True, alpha=.5, color='red', label='Suco Original')
plt.title("Comparao entre o suco de frutas original e o suco de frutas da estao Primavera")
plt.legend()
plt.show()
```
