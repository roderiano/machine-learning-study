```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

# definir uma semente aleatória para garantir a reprodutibilidade dos resultados
np.random.seed(60)

# Criação de dados fictícios para a acidez do suco de frutas em duas estações do ano: Verão e Primavera.
# A acidez do suco de frutas em diferentes estações do ano pode seguir distribuições diferentes.
# Vamos assumir que a acidez do suco de frutas no Verão segue uma distribuição com média 5 e desvio padrão 1,
# enquanto a acidez do suco de frutas na Primavera segue uma distribuição com média -5 e desvio padrão 1.

data_summer = np.random.normal(5, scale=1, size=700)  # 700 pontos com média 5 e desvio padrão 1 para o Verão
data_spring = np.random.normal(-5, scale=1, size=300)  # 300 pontos com média -5 e desvio padrão 1 para a Primavera

data = np.concatenate([data_summer, data_spring])  # Combina os dados de Verão e Primavera

# Visualiza o conjunto de dados original com um histograma
plt.hist(data, bins=30, density=True)
plt.title("Acidez do suco de frutas original")
plt.show()

# Cria e treina um modelo de Mistura Gaussiana com dois componentes (dois picos) para modelar a acidez do suco de frutas
gmm = GaussianMixture(n_components=2)
gmm.fit(data.reshape(-1, 1))  # Ajusta o modelo aos dados

# Usa o modelo treinado para gerar 1000 novos pontos de dados simulando o suco de frutas da estação Primavera
samples = gmm.sample(1000)
sample_spring = samples[0]

# Plota um histograma comparando a acidez do suco de frutas original e a acidez do suco de frutas simulado para a estação Primavera
plt.hist(sample_spring, bins=30, density=True, alpha=.5, color='green', label='Suco Primavera')
plt.hist(data, bins=30, density=True, alpha=.5, color='red', label='Suco Original')
plt.title("Comparação entre o suco de frutas original e o suco de frutas da estação Primavera")
plt.legend()
plt.show()
```
