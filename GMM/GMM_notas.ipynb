```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

# Suponha que temos dados de desempenho de estudantes em duas disciplinas: Matemática e Ciências.
# Vamos gerar dados de exemplo representando o desempenho de estudantes nessas duas disciplinas.

# Gerar dados artificiais para a disciplina de Matemática
np.random.seed(0)
mean_math = [70, 70]  # Média das notas de Matemática
cov_math = [[100, 20], [20, 100]]  # Matriz de covariância das notas de Matemática
math_scores = np.random.multivariate_normal(mean_math, cov_math, 100)

# Gerar dados artificiais para a -disciplina de Ciências
mean_science = [65, 75]  # Média das notas de Ciências
cov_science = [[100, 10], [10, 100]]  # Matriz de covariância das notas de Ciências
science_scores = np.random.multivariate_normal(mean_science, cov_science, 100)

# Combinar os dados de Matemática e Ciências em um único conjunto
data = np.concatenate((math_scores, science_scores))

# Criar e ajustar um modelo GMM com 2 componentes
gmm = GaussianMixture(n_components=2)
gmm.fit(data)

# Prever as probabilidades de pertencer a cada componente (nesse caso, cada disciplina)
probs = gmm.predict_proba(data)

# Plotar os dados e as distribuições encontradas pelo GMM
plt.scatter(data[:, 0], data[:, 1], c=probs[:, 0], cmap='viridis', s=50, edgecolors='k')
plt.scatter(gmm.means_[:, 0], gmm.means_[:, 1], marker='x', s=100, c='red')
plt.title('Distribuições estimadas pelo GMM - Desempenho de Estudantes')
plt.xlabel('Nota em Matemática')
plt.ylabel('Nota em Ciências')
plt.colorbar(label='Probabilidade de pertencer à disciplina de Matemática')
plt.show()
```
