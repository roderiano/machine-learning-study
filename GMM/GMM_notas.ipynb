```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

# Suponha que temos dados de desempenho de estudantes em duas disciplinas: Matemtica e Cincias.
# Vamos gerar dados de exemplo representando o desempenho de estudantes nessas duas disciplinas.

# Gerar dados artificiais para a disciplina de Matemtica
np.random.seed(0)
mean_math = [70, 70]  # Mdia das notas de Matemtica
cov_math = [[100, 20], [20, 100]]  # Matriz de covarincia das notas de Matemtica
math_scores = np.random.multivariate_normal(mean_math, cov_math, 100)

# Gerar dados artificiais para a -disciplina de Cincias
mean_science = [65, 75]  # Mdia das notas de Cincias
cov_science = [[100, 10], [10, 100]]  # Matriz de covarincia das notas de Cincias
science_scores = np.random.multivariate_normal(mean_science, cov_science, 100)

# Combinar os dados de Matemtica e Cincias em um nico conjunto
data = np.concatenate((math_scores, science_scores))

# Criar e ajustar um modelo GMM com 2 componentes
gmm = GaussianMixture(n_components=2)
gmm.fit(data)

# Prever as probabilidades de pertencer a cada componente (nesse caso, cada disciplina)
probs = gmm.predict_proba(data)

# Plotar os dados e as distribuies encontradas pelo GMM
plt.scatter(data[:, 0], data[:, 1], c=probs[:, 0], cmap='viridis', s=50, edgecolors='k')
plt.scatter(gmm.means_[:, 0], gmm.means_[:, 1], marker='x', s=100, c='red')
plt.title('Distribuies estimadas pelo GMM - Desempenho de Estudantes')
plt.xlabel('Nota em Matemtica')
plt.ylabel('Nota em Cincias')
plt.colorbar(label='Probabilidade de pertencer  disciplina de Matemtica')
plt.show()
```
